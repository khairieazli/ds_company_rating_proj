```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df = pd.read_csv('glassdoor_scraper.csv')
```

```python
df
```

```python
#Company rating parsing
df = df[df['Rating'] != -1]
```

```python
#Get company name text only
df['Company Name'] = df['Company Name'].apply(lambda x : x.split('\r')[0])
```

```python
#See location of jobs, also to check for outliers
df['Location'].value_counts()
```

```python
#Age of company
df['Age'] = df.Founded.apply(lambda x : x if x <1 else 2020 - x)
```

```python
#Parsing job description(python, etc.)
```

```python
#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.python_yn.value_counts()
```

```python
#sql
df['sql_yn'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df.sql_yn.value_counts()
```

```python
#tableau
df['tableau_yn'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df.tableau_yn.value_counts()
```

```python
#spark
df['spark_yn'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark_yn.value_counts()
```

```python
#aws
df['aws_yn'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws_yn.value_counts()
```

```python
#excel
df['excel_yn'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel_yn.value_counts()
```

```python
df.drop(['Salary Estimate', 'Headquarters', 'Competitors'], axis=1, inplace=True)
```

```python
df.to_csv('company_rating_data_clean.csv',index=False)
```
